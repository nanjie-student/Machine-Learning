In this notebook we are plotting four concentric circles and explain the idea of variance and bias.  

```python
import matplotlib.pyplot as plt
%matplotlib inline
import random

fig, ax = plt.subplots(2,2)
circle1 = plt.Circle((0.5,0.5), 0.5, fill = False)
ax[0,0].set_aspect(1)
ax[0,0].add_patch(circle1)

circle2 = plt.Circle((0.5,0.5), 0.2, fill = False)
ax[0,0].set_aspect(1)
ax[0,0].add_patch(circle2)

circle3 = plt.Circle((0.5,0.5), 0.3, fill = False)
ax[0,0].set_aspect(1)
ax[0,0].add_patch(circle3)

circle4 = plt.Circle((0.5,0.5), 0.1, color = 'red')
ax[0,0].set_aspect(1)
ax[0,0].add_patch(circle4)

ax[0][1].set_title('Low Variance')
x = []
y = []

for i in range(0, 20):
    temp1 = round(random.uniform(0.50, 0.60), 2)
    temp2 = round(random.uniform(0.50, 0.60), 2)
    x.append(temp1)
    y.append(temp2)

ax[0][0].scatter(x,y);

circle11 = plt.Circle((0.5,0.5), 0.5, fill = False)
ax[0,1].set_aspect(1)
ax[0,1].add_patch(circle11)

circle21 = plt.Circle((0.5,0.5), 0.2, fill = False)
ax[0,1].set_aspect(1)
ax[0,1].add_patch(circle21)

circle31 = plt.Circle((0.5,0.5), 0.3, fill = False)
ax[0,1].set_aspect(1)
ax[0,1].add_patch(circle31)

circle41 = plt.Circle((0.5,0.5), 0.1, color = 'red')
ax[0,1].set_aspect(1)
ax[0,1].add_patch(circle41)

ax[0][0].set_title('Low Variance')
ax[0][0].set_ylabel('low Bias')

x = []
y = []
# Set a length of the list to 10
for i in range(0, 20):
    temp1 = round(random.uniform(0.30, 0.80), 2)
    temp2 = round(random.uniform(0.30, 0.80), 2)
    x.append(temp1)
    y.append(temp2)

ax[0][1].scatter(x,y);


circle11 = plt.Circle((0.5,0.5), 0.5, fill = False)
ax[1,0].set_aspect(1)
ax[1,0].add_patch(circle11)

circle21 = plt.Circle((0.5,0.5), 0.2, fill = False)
ax[1,0].set_aspect(1)
ax[1,0].add_patch(circle21)

circle31 = plt.Circle((0.5,0.5), 0.3, fill = False)
ax[1,0].set_aspect(1)
ax[1,0].add_patch(circle31)

circle41 = plt.Circle((0.5,0.5), 0.1, color = 'red')
ax[1,0].set_aspect(1)
ax[1,0].add_patch(circle41)
ax[1][0].set_ylabel('High Bias')


x = []
y = []
# Set a length of the list to 10
for i in range(0, 20):
    temp1 = round(random.uniform(0.65, 0.70), 2)
    temp2 = round(random.uniform(0.65, 0.70), 2)
    x.append(temp1)
    y.append(temp2)
ax[1][0].scatter(x,y);


circle11 = plt.Circle((0.5,0.5), 0.5, fill = False)
ax[1,1].set_aspect(1)
ax[1,1].add_patch(circle11)

circle21 = plt.Circle((0.5,0.5), 0.2, fill = False)
ax[1,1].set_aspect(1)
ax[1,1].add_patch(circle21)

circle31 = plt.Circle((0.5,0.5), 0.3, fill = False)
ax[1,1].set_aspect(1)
ax[1,1].add_patch(circle31)

circle41 = plt.Circle((0.5,0.5), 0.1, color = 'red')
ax[1,1].set_aspect(1)
ax[1,1].add_patch(circle41)



x = []
y = []
# Set a length of the list to 10
for i in range(0, 30):
    temp1 = round(random.uniform(0.50, 0.80), 2)
    temp2 = round(random.uniform(0.70, 1.20), 2)
    x.append(temp1)
    y.append(temp2)

ax[1][1].scatter(x,y);
```
